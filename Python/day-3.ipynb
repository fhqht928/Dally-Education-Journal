```python
# range(start_num,end_num,step_num) : start_num 부터 end_num-1까지 step_num만큼 증감하여 숫자를 반복
#range(10) => start_num : 0, step num : 1

array = [273,45, 103, 54, 101]
for i in range(len(array)):
    print("{}번째 데이터 : {}".format(i, array[i]))
    
#뒤에서부터 출력 하세요.

for i in range(len(array)-1,0-1,-1):
    print("{}번째 데이터 : {}".format(i, array[i]))
    
#reverse() : 거꾸로 가져오는 함수
for i in reversed(range(len(array))):
    print("{}번째 데이터 : {}".format(i, array[i]))

for item in reversed(array):
    print(item)
```

```python
#키보드에서 이름을 입력받아 리스트에 저장한 후 (이름에 'q'가 입력되면 입력 종료)
#검색하고자 하는 이름을 입력받아 몇 번째 입력된 이름인지 출력하세요.
#enumerate 함수를 이용

list_a = []

while True:
    input_a = input("리스트에 저장할 것들을 작성하시오.\n")
    
    if input_a== 'q':
        break
    list_a.append(input_a)
    
input_b = input("리스트에 저장한 이름을 작성하시오\n")
for i, name in enumerate(list_a):
    if name == input_b:
        print("{}번째 데이터는 {} 입니다.".format(i+1,name))
        break
#입력된 이름의 갯수만큼 점수를 입력하여 score[] 리스트에 저장하세요.
#저장된 이름과 점수를 출력 : 홍길동 90 김철수 100
#딕셔너리로 변화해서 출력


score = []
i = 0
while i < len(list_a):
    score.append(int(input("점수 입력.\n")))
    i+=1
    print(len(list_a))
    
print("list_a[] : score[]")
for idx in range(len(list_a)):
    print("{} : {}".format(list_a[idx], score[idx]))

dict_score = {}
i = 0
while i < len(list_a) : 
    dict_score[list_a[i]] = score[i]
    i+=1
print("dict_score : ",)
for key, value in dict_score.items():
    print("{} : {}".format(key, value))
```

```python
#키보드에서 입력받는 함수 : 숫자와 부호를 입력받아 list로 반환하는 함수, 입력의 끝은 'q'이면 종료
# 부호가 '+'이면 합을 구하는 함수로
# 부호가 '-'이면 차를 구하는 함수로

def input_data():
    input_list = input("10 + 20 의 형식으로 입력").split()
    return input_list

    
def plus_func(num1,num2):
    return num1 + num2
def minus_func(num1,num2):
    return num1 - num2
```

```python
while True:
    input_list = input_data()
    if 'q' in input_list:
        break
    if input_list[1] == '+':
        result = plus_func(int(input_list[0]),int(input_list[2]))
    elif input_list[1] == '-':
        result = minus_func(int(input_list[0]),int(input_list[2]))
    print("{} {} {} = {}".format(input_list[0], input_list[1],input_list[2],result))
```
